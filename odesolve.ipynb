```python
import numpy as np
import matplotlib.pyplot as plt
from odesolve import odesolve

def f(X, t):
    x1, x2 = X
    dx1dt = x2
    dx2dt = -x1
    dXdt = [dx1dt, dx2dt]
    return np.array(dXdt)

#Initial confitions
x0 = 1
y0 = 0
X0 = np.array([x0, y0])
h = 0.01                          # max step size
t = np.linspace(0, 10, 100)       # times to plot the solution

Xt = odesolve(f, X0, t, h)

plt.plot(t, Xt)
plt.savefig('shm.pdf')
plt.show()
```
